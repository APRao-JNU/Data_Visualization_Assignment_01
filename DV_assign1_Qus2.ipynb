# Question 2. Nitin had visited a zoo recently and had collected the following data. How can Nitin use a scatter plot to represent this data?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_frame = pd.read_csv("Qus2_data.csv")
```

```python
data_frame
```

```python
x_axis=data_frame["Type of Animal"]
```

```python
x_axis
```

```python
y_axis=data_frame["Number of Animals in the Zoo"]
```

```python
y_axis
```

## Scatter Plot 

```python
plt.scatter(x_axis,y_axis,marker="D")
plt.xlabel("Type of Animal")
plt.ylabel("Number of Animals")
plt.yticks(np.arange(0, max(y_axis)+1, 5))
plt.title("Type of Animal VS Number of Animal")
plt.show()
```

## Line Plot

```python
plt.plot(x_axis,y_axis,"b") 
plt.xlabel("Type of Animal")
plt.ylabel("Number of Animals")
plt.yticks(np.arange(0, max(y_axis)+1, 5))
plt.title("Type of Animal VS Number of Animal")
plt.show()
```

## Scatter + Line Plot

```python
plt.plot(x_axis,y_axis,"b") 
plt.scatter(x_axis,y_axis,marker="D",c="red")
plt.xlabel("Type of Animal")
plt.ylabel("Number of Animals")
plt.yticks(np.arange(0, max(y_axis)+1, 5))
plt.title("Type of Animal VS Number of Animal")
plt.show()
```

## Step Plot

```python
plt.scatter(x_axis,y_axis,marker="8",s=85)
plt.step(x_axis,y_axis,"purple") 
plt.xlabel("Type of Animal")
plt.ylabel("Number of Animals")
plt.yticks(np.arange(0, max(y_axis)+1, 5))
plt.title("Type of Animal VS Number of Animal")
plt.show()
```

## Plotting with Grid

```python
plt.plot(x_axis,y_axis,"r") 
plt.scatter(x_axis,y_axis,marker="D",c="green")
plt.xlabel("Type of Animal")
plt.ylabel("Number of Animals")
plt.grid()
plt.yticks(np.arange(0, max(y_axis)+1, 5))
plt.title("Type of Animal VS Number of Animal")
plt.show()
```
