# Question 3. The meteorological department has collected the following data about the temperature and humidity in their town. Refer to the table given below and indicate the method to find the humidity at a temperature of 60 degrees Fahrenheit.

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("Qus3_data.csv")
data_frame
```

```python
x_axis = data_frame["Temperature (Degree Fahrenheit)"]
```

```python
x_axis
```

```python
y_axis = data_frame["Humidity(%)"]
```

```python
y_axis
```

### First let's plot the graph and scatter the points to visualize the given data.

```python
plt.scatter(x_axis,y_axis,marker="o")
plt.xlabel("Temperature (Degree Fahrenheit)")
plt.ylabel("Humidity(%)")
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.title("Temperature VS Humidity")
plt.show()
```

```python
plt.plot(x_axis,y_axis,"y") 
plt.xlabel("Temperature (Degree Fahrenheit)")
plt.ylabel("Humidity(%)")
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.title("Temperature VS Humidity")
plt.show()
```

```python
plt.scatter(x_axis,y_axis,marker="D",s=85,c="lime")
plt.step(x_axis,y_axis,"gray")
plt.xlabel("Temperature (Degree Fahrenheit)")
plt.ylabel("Humidity(%)")
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.title("Temperature VS Humidity")
plt.show()
```

```python
import seaborn as sns
sns.lmplot(x="Temperature (Degree Fahrenheit)", y="Humidity(%)", data=data_frame)
plt.title("Temperature VS Humidity")
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.show()
```

```python
plt.plot(x_axis,y_axis,"blue") 
plt.scatter(x_axis,y_axis,marker="o",c="purple")
plt.xlabel("Temperature (Degree Fahrenheit)")
plt.ylabel("Humidity(%)")
plt.grid()
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.title("Temperature VS Humidity")
plt.show()
```

# So i guess based on given information can be 60 degree farenhiet between 45 to 50%.

# If we checking by vertical Lines

```python
plt.plot(x_axis,y_axis,"green",linestyle='dotted') 

for i in x_axis:
    plt.axvline(i)

plt.scatter(x_axis,y_axis,marker="o",c="red")
plt.xlabel("Temperature (Degree Fahrenheit)")
plt.ylabel("Humidity(%)")
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.title("Temperature VS Humidity")
plt.show()
```

# By Horizontal same as vertical lines

```python
plt.plot(x_axis,y_axis,"green",linestyle='dotted') 

for i in x_axis:
    plt.axvline(i)

for j in y_axis:
    plt.axhline(j)

plt.scatter(x_axis,y_axis,marker="o",c="red")
plt.xlabel("Temperature (Degree Fahrenheit)")
plt.ylabel("Humidity(%)")
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.title("Temperature VS Humidity")
plt.show()
```

# And by Horizontal as well as vertical lines (with grid enabled)

```python
plt.plot(x_axis,y_axis,"green",linestyle='dotted') 

for i in x_axis:
    plt.axvline(i)

for j in y_axis:
    plt.axhline(j)

plt.scatter(x_axis,y_axis,marker="o",c="red")
plt.xlabel("Temperature (Degree Fahrenheit)")
plt.ylabel("Humidity(%)")
plt.grid()
plt.xticks(np.arange(0, max(x_axis)+1, 15))
plt.yticks(np.arange(0, max(y_axis)+1, 10))
plt.title("Temperature VS Humidity")
plt.show()
```
