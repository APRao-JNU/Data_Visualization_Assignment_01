# Question 1. The following scatter plot excel data for age (of the child in years) and height (of the child in feet) can be represented as a scatter plot in excel.

### a) Indentify the x-axis and y-axis for the scatter plot.
### b) Define the scale for each of the axes.
### c) Plot the points based on their values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
!pip install seaborn
```

```python
data_frame = pd.read_csv("Qus1_data.csv")
data_frame
```

```python
x_axis = data_frame["Age of the Child"]
```

```python
x_axis
```

```python
y_axis = data_frame["Height"]
```

```python
y_axis
```

## SCATTER PLOT

```python
plt.scatter(x_axis,y_axis)
plt.xlabel("Age of the child (in Years)")
plt.ylabel("Height of the child (in feet)")
plt.xticks(np.arange(0, max(x_axis)+1, 1.0))
plt.yticks(np.arange(0, max(y_axis)+1, 0.5))
plt.title("AGE VS HEIGHT DATA")
plt.show()
```

## LINE PLOT

```python
plt.plot(x_axis,y_axis,"black") # 'r','o'/'orange' is the color red
plt.xlabel("Age of the child (in Years)")
plt.ylabel("Height of the child (in feet)")
plt.xticks(np.arange(0, max(x_axis)+1, 1.0))
plt.yticks(np.arange(0, max(y_axis)+1, 0.5))
plt.title("AGE VS HEIGHT DATA")
plt.show()
```

## Scatter with Line Plot 

```python
plt.scatter(x_axis,y_axis,marker="o",s=80)
plt.plot(x_axis,y_axis,"yellow") 
plt.xlabel("Age of the child (in Years)")
plt.ylabel("Height of the child (in feet)")
plt.xticks(np.arange(0, max(x_axis)+1, 1.0))
plt.yticks(np.arange(0, max(y_axis)+1, 0.5))
plt.title("AGE VS HEIGHT DATA")
plt.show()
```

## Plotting with a grid

```python
plt.scatter(x_axis,y_axis,marker="o",s=80)
plt.plot(x_axis,y_axis,"red") # 'r','o'/'orange' is the color red
plt.grid()
plt.xlabel("Age of the child (in Years)")
plt.ylabel("Height of the child (in feet)")
plt.xticks(np.arange(0, max(x_axis)+1, 1.0))
plt.yticks(np.arange(0, max(y_axis)+1, 0.5))
plt.title("AGE VS HEIGHT DATA")
plt.show()
```

```python
import seaborn as sns
sns.lmplot(x="Age of the Child", y="Height", data=data_frame)
plt.title("AGE VS HEIGHT DATA")
plt.xticks(np.arange(0, max(x_axis)+1, 1.0))
plt.yticks(np.arange(0, max(y_axis)+1, 0.5))
plt.show()
```
